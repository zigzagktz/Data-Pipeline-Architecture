```python
from  selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
```

```python

# https://sites.google.com/a/chromium.org/chromedriver/home 
# link to get driver

url = "https://www.nordstromrack.com/shop/Men/Shoes"

driver = webdriver.Chrome()
driver.implicitly_wait(30)
driver.get(url)

colname = ["Item","Retail Price","Selling Price","Discount"]

mydf= pd.DataFrame()

try:
    for i in range(0,5):
        driver.find_element_by_link_text("Next").click()    
        soup=BeautifulSoup(driver.page_source, 'lxml')


        lis = []
        for i in (soup.find_all("div",class_="product-grid-item__details")):
            lis.append(i.find_all(text=True)


        second_lis = []
        for i in lis:
            second_lis.append(i.split("$"))

        for i in range(0,len(second_lis)-1):
            if len(second_lis[i])==3:
                empt = second_lis[i][len(second_lis[i])-1]
                second_lis[i][2] = empt[0:5]
                second_lis[i].append(empt[5:])
                structured = pd.DataFrame(second_lis,columns=colname)

        del empt
        second_lis.clear()
        mydf = mydf.append(structured)
        print("loop ran")

except:
    driver.close()
    
driver.close()
```
