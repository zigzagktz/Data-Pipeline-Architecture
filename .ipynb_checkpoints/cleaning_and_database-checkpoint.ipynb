```python
import boto3
import pandas as pd
import json
#import sqlalchemy
import psycopg2
```

```python
def formating(data):
  subset = {"subBrandName","beautyCategoryName","productName","reviewHoverMessage ","averageRating","priceInfo"}
  sub_lis = []
  for i in data["products"]:
      item = i["productInfo"]
      item = {key:value for key,value in item.items() if key in subset}        
      sub_lis.append(item)
  return sub_lis


def cleaning_api(sub_lis):
  subset_price= {"regularPrice",'unitPrice'}
    
  for i in range(0,len(sub_lis)):
      item = sub_lis[i]["priceInfo"]
      item = {key:value for key,value in item.items() if key in subset_price}
      sub_lis[i].pop('priceInfo')
      sub_lis[i].update(item)
      sub_lis[i]["averageRating"] = float(sub_lis[i]["averageRating"])
      

      if 'regularPrice' in sub_lis[i]:
          sub_lis[i]["regularPrice"] =  sub_lis[i]["regularPrice"].split("/")[-1]
          sub_lis[i]["regularPrice"] =  sub_lis[i]["regularPrice"].replace(",","")
          sub_lis[i]['regularPrice'] = float(sub_lis[i]["regularPrice"].replace("$",""))
      else:
          pass

      if 'unitPrice' in sub_lis[i]:
          sub_lis[i]['unitPrice'] =  sub_lis[i]['unitPrice'].split("/")[-1]
          sub_lis[i]['unitPrice'] =  sub_lis[i]['unitPrice'].replace(",","")
          sub_lis[i]['unitPrice'] = float(sub_lis[i]["unitPrice"].replace("$",""))
      else:
          pass    
        
  df = pd.DataFrame(sub_lis)
  df.replace({pd.np.nan:None})
  return df.values.tolist()
```

```python
def cleaning(mylist2):
  best_seller = []
  sponsored = []
  normal=[]

  for i in mylist2:
    temp = i.replace("\n","$").split("$")
    temp = [i.strip() for i in temp]
    temp = [x for x in temp if x!=""]

    if temp[0]=="Sponsored":
      sponsored.append(temp)
    elif temp[0]=="Best Seller":
      best_seller.append(temp)
    else:
      normal.append(temp)

  for i in range(0,len(sponsored)):
    sponsored[i].pop(0)
    sponsored[i].pop(0)
    sponsored[i].pop(0)
    sponsored[i] = sponsored[i][0:7]


  for i in range(0,len(best_seller)):
    if best_seller[i][2]=="Sponsored":
      best_seller[i].pop(0)
      best_seller[i].pop(0)
      best_seller[i].pop(0)
      best_seller[i].pop(0)
      best_seller[i].pop(0)
      best_seller[i] = best_seller[i][0:7]
    else:    
      best_seller[i].pop(0)
      best_seller[i].pop(0)
      best_seller[i] = best_seller[i][0:7]

  for i in range(0,len(normal)):
    normal[i] = normal[i][0:7]
    
  normal = [x for x in normal if len(x)==7  ]
  sponsored = [x for x in sponsored if len(x)==7  ]
  best_seller = [x for x in best_seller if len(x)==7  ]
  
  normal.extend(sponsored)
  normal.extend(best_seller)
    
  for i in range(0,len(normal)):
    normal[i][2] = float(normal[i][2].split()[0])
    normal[i][3]= normal[i][3].replace(",","")
    normal[i].pop(4)
    try:
      normal[i][4] = float(normal[i][4])
    except:
      normal[i][4] = None
    
    try:
      normal[i][-1] = float(normal[i][-1])
    except:
      normal[i][-1] = None

  
  
  return normal

```

```python
def conversion(mylist):
  for i in range(0,len(mylist)):
    mylist[i][-1] = "".join([item for item in mylist[i][-1] if item.isdigit()==True])
  return mylist  
```

```python
key = 
secret_key = 

def connect(key, secret_key):
  region = "us-east-1"
  aws_access_key_id = key
  aws_secret_access_key = secret_key
  s3 = boto3.client("s3", region_name = region,aws_access_key_id= aws_access_key_id, aws_secret_access_key= aws_secret_access_key )
  return s3

def fetching(s3, bucket_name, key ):
  obj = s3.get_object(Bucket=bucket_name,Key= key)
  mylist = obj["Body"].read()
  return json.loads(mylist.decode("utf-8"))



def databases():
  host = "database-1.cdqvm7ggjees.us-east-1.rds.amazonaws.com"
  port = 5432
  databasename = "postgres"
  user = "postgres"
  password = "password"
  conn = psycopg2.connect(host = host, user=user, port=port, password=password , database=databasename)
  cur = conn.cursor()
  return cur


def create_tables(cur):
  cur.execute("""rollback""")
  cur.execute("""DROP TABLE IF EXISTS Website1""")
  cur.execute(""" CREATE TABLE Website1 
                    (
                    Brand varchar(255),
                    Item varchar(255),
                    Retail_Price  money ,
                    Selling_Price  money ,
                    Discount int );
                    """)
  
  cur.execute("""DROP TABLE IF EXISTS Website2""")
  cur.execute(""" CREATE TABLE Website2 
                    (
                    Brand varchar(255),
                    Item varchar(255),
                    Out_of_5 int,
                    Num_of_votes int,
                    Selling_Price  int ,
                    Retail_Price  int );  """)
  cur.execute("""commit""")
  
  cur.execute("""DROP TABLE IF EXISTS Website3""")
  cur.execute(""" CREATE TABLE Website3
                    (
                    Rating float,
                    Brand varchar(255),
                    Product_name varchar(255),
                    Regular_Price float ,
                    Brand_2 varchar(255) ,
                    Unit_Price  float );  """)
  cur.execute("""commit""")
  
  
  

def write_DB(cur, mylist):
  cur.execute("""rollback""")
  for i in mylist:
    if len(i) ==5:
      cur.execute(""" INSERT INTO website1 values (%s,%s,%s,%s,%s)  """, tuple(i))
  cur.execute("""commit""")

def write_DB2(cur, mylist):
  cur.execute("""rollback""")
  for i in mylist:
    cur.execute(""" INSERT INTO website2 values (%s,%s,%s,%s,%s,%s)  """, tuple(i))
  cur.execute("""commit""")
  
def write_DB3(cur, mylist):
  cur.execute("""rollback""")
  for i in mylist:
    cur.execute(""" INSERT INTO website3 values (%s,%s,%s,%s,%s,%s)  """, tuple(i))
  cur.execute("""commit""")

 
      
def web1(bucket_name,filename):
  s3 = connect(key,secret_key)
  mylist = fetching(s3,bucket_name,filename)
  mylist = conversion(mylist)  
  cur = databases()
  create_tables(cur)
  write_DB(cur, mylist)
  return mylist

def web2(bucket_name,filename):
  s3 = connect(key,secret_key)
  mylist = fetching(s3,bucket_name,filename)
  mylist = cleaning(mylist)  
  cur = databases()
#  create_tables(cur)
  write_DB2(cur, mylist)
  return mylist

def web3(bucket_name,filename):
  s3 = connect(key,secret_key)
  mylist = fetching(s3,bucket_name,filename)
  mylist = formating(mylist)
  mylist = cleaning_api(mylist)
  cur = databases()
#  create_tables(cur)
  write_DB3(cur, mylist)
  return mylist
  


mylist = web1("mycurrentproject","myfile.txt")
mylist2 = web2("mycurrentproject","myfile2.txt")
mylist3 = web3("mycurrentproject","myfile3.txt")
```
