```python
def read_json():
    def get_data():
        url = "https://walgreens.p.rapidapi.com/products/list"
        querystring = {"sort":"Newest Arrival","p":"1","id":"359432","s":"3600"}
        headers = {
            'x-rapidapi-host': "walgreens.p.rapidapi.com",
            'x-rapidapi-key': "625d329734mshc974410aaf2549cp1879aejsncff2ccd408d8"
            }
        response = requests.request("GET", url, headers=headers, params=querystring)
        return response.text
    
    data = json.loads(get_data())
    
    
    
    subset = {"subBrandName","beautyCategoryName","productName","reviewHoverMessage ","averageRating","priceInfo"}
    sub_lis = []

    for i in data["products"]:
        item = i["productInfo"]
        item = {key:value for key,value in item.items() if key in subset}        
        sub_lis.append(item)

    return sub_lis

def cleaning(sub_lis):
    
    subset_price= {"regularPrice",'unitPrice'}
    
    for i in range(0,len(sub_lis)):
        item = sub_lis[i]["priceInfo"]
        item = {key:value for key,value in item.items() if key in subset_price}
        sub_lis[i].pop('priceInfo')
        sub_lis[i].update(item)
        sub_lis[i]["averageRating"] = float(sub_lis[i]["averageRating"])
        
        if 'regularPrice' in sub_lis[i]:
            sub_lis[i]['regularPrice'] = sub_lis[i]["regularPrice"].replace("$","")
        else:
            pass
        
        if 'unitPrice' in sub_lis[i]:
            sub_lis[i]['unitPrice'] = sub_lis[i]["unitPrice"].replace("$","")
        else:
            pass    
    return sub_lis

cleaning(read_json())
    
```
